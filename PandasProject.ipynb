```python
len(hotels)
```

**Ищем пропущенные значения, точнее их количество**

```python
hotels.isnull().sum()
```

```python
print(f"В колонке company значения отсутствуют для {hotels['company'].isna().sum()} строк.")
```

**Удаляем колонку company, поскольку она имеет слишком большое количество отсутствующих строк**

```python
hotels = hotels.drop('company',axis=1)
```

**Ищу самые популярные страны в датасете**

```python
hotels['country'].value_counts()[:5]
```

**Ищем человека который заплатил наибольшее количество денег за сутки, также выводим это значение**

```python
hotels.sort_values('adr',ascending=False)[['adr','name']].iloc[0]
```

**Ищем среднее значение adr для всех бронирований**

```python
round(hotels['adr'].mean(),2)
```

**Среднее количество ночей, усредненное по всем бронированиям**

```python
hotels['total_stay_days'] = hotels['stays_in_week_nights'] + hotels['stays_in_weekend_nights']

round(hotels['total_stay_days'].mean(),2)
```

**Средняя общая сумма бронирования в этом наборе данных**

```python
hotels['total_paid'] = hotels['adr'] * hotels['total_stay_days']
round(hotels['total_paid'].mean(),2)
```

**Выводим имена и адреса email тех, кто сделал ровно 5 специальных запросов**

```python
hotels[hotels['total_of_special_requests'] == 5][['name','email']]
```

**Процент бронирований классифицированы как бронирования "is_repeated_guest"?**

```python
round(100 * sum(hotels['is_repeated_guest'] == 1) / len(hotels),2)
```

**5 фамилий встречаются в этом наборе данных наиболее часто**

```python
hotels['name'].apply(lambda name: name.split()[1]).value_counts()[:5]
```

**Имена людей, которые бронировали номер для наибольшего количества детей и младенцев(колонки [babies, children]**

```python
hotels['total_kids'] = hotels['babies'] + hotels['children']

hotels.sort_values('total_kids',ascending=False)[['name','adults','total_kids','babies','children']][:3]
```

**3 кода области котоые встречаются в наборе данных наиболее часто?**

```python
print('Code - Total Count')
hotels['phone-number'].apply(lambda num:num[:3]).value_counts()[:3]
```

**Считаю количество заселений в отель между 1м и 15м числами месяца**

```python
hotels['arrival_date_day_of_month'].apply(lambda day: day in range(1,16)).sum()
```


**Создаю таблицу, в которой будет количество заселений по дням недели.**

```python
import numpy as np
```

```python
def convert(day,month,year):
    return f'{day}-{month}-{year}'
```

```python
hotels['date']  = np.vectorize(convert)(hotels['arrival_date_day_of_month'],
                                        hotels['arrival_date_month'],
                                        hotels['arrival_date_year'])
```

```python
hotels['date'] = pd.to_datetime(hotels['date'])
```

```python
hotels['date'].dt.day_name().value_counts()
```
